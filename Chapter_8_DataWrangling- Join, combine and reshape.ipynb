## Database-style DataFrame joins

```python
df1=pd.DataFrame({'key':['b','b','a','c','a','a','b'],'data1':range(7)})
df2=pd.DataFrame({'key':['a','b','d'],'data2':range(3)})
df1
```

```python
df2
```

```python
df1.merge(df2) #pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on='key')
```

```python
df3=pd.DataFrame({'lkey':['b','b','a','c','a','a','b'],'data1':range(7)})
df4=pd.DataFrame({'rkey':['a','b','d'],'data2':range(3)})
```

```python
pd.merge(df1,df2,left_on='lkey',right_on='rkey')
```

```python
pd.merge(df1,df2,how='inner')
```

```python
pd.merge(df1,df2,how='outer')
```

```python
df3=pd.DataFrame({'key':['b','b','a','c','a','b'],'data1':range(6)})
df4=pd.DataFrame({'key':['a','b','a','b','d'],'data2':range(5)})
print(df3)
df4
```

```python
pd.merge(df3,df4,how='left')
```

```python
pd.merge(df3,df4,how='inner')
```

```python
df3=pd.DataFrame({'key1':['foo','foo','bar'],'key2':['one','two','one'],'lval':[1,2,3]})
df4=pd.DataFrame({'key1':['foo','foo','bar','bar'],'key2':['one','one','one','two'],'rval':[4,5,6,7]})
df3
```

```python
df4
```

```python
pd.merge(df3,df4,on=['key1','key2'],how='outer')
```

```python
pd.merge(df3,df4, on='key1') # overlapping col names automatically appended by _x and _y
```

```python
#we can do it manually
pd.merge(df3,df4,on='key1',suffixes=('_left','_right'))
```

```python
pd.merge(df3,df4,on='key1',suffixes=('_left','_right'))
```

### Merginge on Index

```python
left=pd.DataFrame({'key':['a','b','a','a','b','c'],'data1':range(6)})
right=pd.DataFrame({'group_val':[3.5,7]},index=['a','b'])
print(left)
right
```

```python
pd.merge(left,right,left_on='key',right_index=True)
```

```python
pd.merge(left,right,left_on='key',right_index=True,how='outer')
```

```python
left1=pd.DataFrame({'key1':['Ohio','Ohio','Ohio','Nevada','Nevada'],'key2':[2000,2001,2002,2001,2002],'data':np.arange(5.0)})
right1=pd.DataFrame(np.arange(12).reshape((6,2)),index=[['Nevada','Nevada','Ohio','Ohio','Ohio','Ohio'],[2001,2000,2000,2000,2001,2002]],columns=['event1','event2'])
print(left1)
right1
```

```python
pd.merge(left,right,left_on=['key1','key2'],right_index=True)
```

```python
pd.merge(left1,right1,left_on=['key1','key2'],right_index=True,how='outer')
```

```python
left2=pd.DataFrame([[1.,2.],[3.,4.],[5.,6.]],index=['a','b','c'],columns=['Ohio','Nevada'])
right2=pd.DataFrame([[7.,8.],[9.,10.],[11.,12.],[13,14]],index=['b','c','d','e'],columns=['missouri','Alabama'])
left2
```

```python
right2
```

```python
pd.merge(left2,right2,left_index=True,right_index=True,how='outer')
```

```python
# there is a join instatnce which can do the same job
left2.join(right2,how='outer')
```

```python
left.join(right,on='key') #index of a passing dataframe that is 'right'
```

```python
another=pd.DataFrame([[7.,8.],[9.,10.],[11.,12.],[16.,17.]],index=['a','c','e','f'],columns=['New york','Oregon'])
another
```

```python
left2.join([right2,another])
```

```python
left2.join([right2,another],how='outer')
```

## Combining Data with overlap

```python
a=pd.DataFrame([np.nan,2.5, np.nan,3.5,4.5,np.nan],index=['f','e','d','c','b','a'])
b=pd.DataFrame(np.arange(len(a),dtype=np.float64),index=['f','e','d','c','b','a'])
b.iloc[-1,]=np.nan
print(a)
b
```

```python
np.where(pd.isnull(a),b,a)
```

```python
b.combine_first(a)
```

# Reshaping and Pivoting

### Reshaping with hierarchical Indexing

```python
data=pd.DataFrame(np.arange(6).reshape((2,3)),index=pd.Index(['Ohio','Colorado'],name='state'),columns=pd.Index(['one','two','three'],name='number'))
data
```

```python
match=data.stack()
match
```

```python
match.unstack()
```

```python
match.unstack(0)
```

```python
match.unstack('state')
```

```python
df1=pd.DataFrame([0,1,2,3],index=['a','b','c','d'])
df2=pd.DataFrame([4,5,6],index=['c','d','e'])
df1
```

```python
df2
```

```python
data=pd.concat([df1,df2],keys=['one','two'])
data
```

```python
data.unstack()
```

```python
data.unstack().stack()
```

```python
data.unstack().stack(dropna=False) #keeping Na values
```

```python
df=pd.DataFrame({'left':match,'right':match+5},columns=pd.Index(['left','right'],name='side'))
df
```

```python
df.unstack('state')
```

```python
df.unstack('state').stack('side')
```

```python
data=pd.read_csv('H:/data structure program/python_for_everybody/macrodata.csv')
data.head()
```

```python
'''periods=pd.PeriodIndex(year=data.year, quarter=data.quarter, name='date')
columns=pd.Index(['realgdp','infl','unemp'],name='item')
data=data.reindex(columns=columns)
data.index=periods.to_timestamp('D','end')
ldata=data.stack().reset_index().rename(columns={0:'value'})
ldata[:10]
'''
periods = pd.PeriodIndex(year=data.year, quarter=data.quarter, name='date')
periods
```

```python
data = pd.DataFrame(data.to_records(),
                 columns=pd.Index(['realgdp', 'infl', 'unemp'], name='item'),
                 index=periods.to_timestamp('D', 'end'))

data
```

```python
ldata = data.stack().reset_index().rename(columns={0: 'value'})
ldata
```

```python
pivoted=ldata.pivot('date','item','value')
pivoted
```

```python
ldata['value2']=np.random.randn(len(ldata))
ldata[:10]
```

```python
pivoted=ldata.pivot('date','item',['value','value2'])
pivoted[:5]
```

```python
pivoted['value'][:5]
```

```python
#pivot is equivalent to set_index and then unstack method
unstack=ldata.set_index(['date','item']).unstack('item')
unstack[:5]
```

### pivoting 'wide' and 'long' format

```python
df=pd.DataFrame({'key':['foo','bar','baz'],'A':[1,2,3],'B':[4,5,6],'C':[7,8,9]})
df
```

```python
melted=pd.melt(df,'key')
melted
```

```python
reshaped=melted.pivot('key','variable','value')
reshaped
```

```python
reshaped.reset_index()
```

```python
pd.melt(df,id_vars='key',value_vars=['A','B'])
```

```python
pd.melt(df,value_vars=['A','B','C'])
```

```python
pd.melt(df,value_vars=['key','A','B'])
```
