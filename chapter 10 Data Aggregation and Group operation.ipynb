### Quantile and Bucket analysis

```python
df=pd.DataFrame({'data1':np.random.randn(1000),'data2':np.random.randn(1000)})
quantiles=pd.cut(df.data1,4)
quantiles[:5]
```

```python
def get_stats(group):
    return {'min':group.min(),'max':group.max(),'count':group.count(),'mean':group.mean()}
grouped=df.data2.groupby(quantiles)
```

```python
grouped.apply(get_stats).unstack()
```

```python
grouped=pd.qcut(df.data1,10,labels=False)
grouped=df.data2.groupby(grouped)
grouped.apply(get_stats).unstack()
```

```python
states=['saboo','leh','shey','stok','phey','khalsi','tia','ang']
group_key=['East']*4+['west']*4
data=pd.Series(np.random.randn(8),index=states)
data
```

### Random Sampling and permutation

```python
#Hearts, spades, clubs, diamond
suits=['H','S','C','D']
card_val=(list(range(1,11))+[10]*3)*4
base_names=['A']+list(range(2,11))+['J','K','Q']
cards=[]
for suit in ['H','S','C','D']:
    cards.extend(str(num)+suit for num in base_names)
deck=pd.Series(card_val, index=cards)
deck[:14]
```

```python
def draw(deck,n=5):
    return deck.sample(n)
draw(deck)
```

```python
get_suit=lambda card: card[-1]
deck.groupby(get_suit).apply(draw,n=2)
```

```python
deck.groupby(get_suit,group_keys=False).apply(draw,n=2)
```
